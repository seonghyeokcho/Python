```python
# 1. 직접 함수 구현
import numpy as np
import matplotlib.pyplot as plt


# 함수 정의
# ========================================
def function(x):
    return x**2


def differential(f,x):
    h = 1e-10
    return (f(x+h) - f(x)) / h


def lines_equation(f, m, x, a):
    return m*(x-a) + f(a)
# ========================================


# 원함수
x = np.arange(0.0, 20.0, 0.01)
y = function(x)

# 접점
a = 10

# 기울기
gradient = differential(function, a)

# 직선의 방정식
lines = lines_equation(function, gradient, x[500:1500], a)

print(f"x = {a} 에서의 접선의 기울기: {gradient}")

plt.figure(figsize=(10, 15))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, "-k")
plt.plot(x[500:1500], lines, "--r")
plt.scatter(10.0, function(10.0), c="blue")
plt.show()
```

```python
import numpy as np

def function(x, m=1, t=1, b=0):
    print(f"fx = {m}x^{t} + {b}")
    return (m*x)**t + b

def integration(a, b, fx, x, m, t, c):
    Fx = (1/(t+1))*fx(x, m, t+1, c)
    result = Fx**a - Fx**b
    return result

x = np.arange(0, 21, 1)
result = integration(0, 20, function, x, 2, 1, 0)
result.sum()
```

```python
import numpy as np

def function(x, m, t):
    return 1/(t+1) * m * x**(t+1)


def integral(a, b, fx, m, t):
    Fa = fx(a, m, t)
    Fb = fx(b, m, t)
    return Fb - Fa


integration = integral(0, 20, function, 1, 2)
print(integration)
```
