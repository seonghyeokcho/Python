```python
import string
import random
import time
from tqdm.notebook import tnrange


# 함수 정의
# ========================================================================
def input_password_number():
    while True:
        try:
            characters_number = int(input("생성할 비빌번호의 길이를 숫자로 입력하세요: "))
            if characters_number < 8:
                print("강력한 비밀번호 설정을 위해 최소 8자 이상이 필요합니다!")
                continue
            else:
                break
        except:
            print("X X X  failed!!  X X X    비밀번호의 길이로 \"숫자\"만 입력하세요!")
            
    print("initialization completed ! ! !")
    print()
    time.sleep(1)  # 극적 효과
    
    return characters_number


def combinate_password(source):
    created = []

    characters_number = input_password_number()
    
    for i in tnrange(characters_number, desc="combinating..."):
        while True:
            if i < round(characters_number * 0.2):  # 특수문자 20%
                element = source["punctuation"][random.randint(0, len(punctuation) - 1)]
                time.sleep(1)  # 극적 효과
                if element not in created:
                    break
            elif round(characters_number * 0.2) <= i < round(characters_number * 0.5):  # 알파벳 소문자 30%
                element = source["lowercase"][random.randint(0, len(lowercase) - 1)]
                time.sleep(1)  # 극적 효과
                if element.upper() not in created and element.lower() not in created:
                    break
            elif round(characters_number * 0.5) <= i < round(characters_number * 0.7):  # 숫자 20%
                element = source["digits"][random.randint(0, len(digits) - 1)]
                time.sleep(1)  # 극적 효과
                if element not in created:
                    break
            else:  # 알파벳 대문자 30%
                element = source["uppercase"][random.randint(0, len(uppercase) - 1)]
                time.sleep(1)  # 극적 효과
                if element.upper() not in created and element.lower() not in created:
                    break
        
        created.append(element)
        random.shuffle(created)  # 2차 셔플(each iteration)
        
    print("combination completed ! ! !")
    print()
    time.sleep(1)  # 극적 효과
        
    return created


def display_created_password(source):
    created = combinate_password(source)
    password = "".join(created[:])
    
    print("-"*(18+len(password)))
    print(f"your password is: {password}")
    print("-"*(18+len(password)))
# ========================================================================


# 실행문
# ========================================================================
lowercase = list(string.ascii_lowercase)
uppercase = list(string.ascii_uppercase)
digits = list(string.digits)
punctuation = list(string.punctuation)

string_dictionary = {
                    "lowercase":lowercase,
                    "uppercase":uppercase,
                    "digits":digits,
                    "punctuation":punctuation
                    }

for i in string_dictionary.keys():  # 1차 셔플(each case)
    random.shuffle(string_dictionary[i])

display_created_password(string_dictionary)  # 비밀번호 생성
# ========================================================================
```

```python
class PasswordEnigma:
    
    def __init__(self, order="encrypt"):
        self.order = order  # 암호생성:encrypt / 암호해독:decrypt
        
        self.punctuation = ['<', '}', '%', ';', '~', ']', '^', '"', ',', '/', '>', '\\', '(', '{', '`', '+', '|', '$', '@', '#', '&', '.', '-', '!', '*', ':', ')', '?', '[', '_', '=', "'"]
        self.digits = ['3', '6', '5', '2', '8', '0', '4', '9', '1', '7']
        self.lowercase = "abcdefghijklmnopqrstuvwxyz"
        self.uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        
        # 플러그 초기화
        # --------------------------------------------------
        left_plugs = ['Y', 'B', 'L', 'U', 'Q', 'S', 'M', 'W', 'Z', 'R', 'D', 'V', 'N', 'G', 'P', 'A', 'H', 'J', 'T', 'I', 'C', 'F', 'K', 'O', 'E', 'X']
        right_plugs = ['W', 'C', 'G', 'H', 'B', 'D', 'V', 'M', 'S', 'I', 'X', 'O', 'E', 'A', 'R', 'U', 'N', 'F', 'P', 'L', 'T', 'Z', 'Y', 'K', 'Q', 'J']
        self.plug_codes_dict1 = dict(zip(left_plugs, right_plugs))
        self.plug_codes_dict2 = dict(zip(right_plugs, left_plugs))
        # --------------------------------------------------
        
        # 회전자 초기화
        # --------------------------------------------------
        self.sequence1_1 = ['B', 'P', 'X', 'Z', 'S', 'G', 'T', 'E', 'L', 'D', 'U', 'A', 'V', 'N', 'C', 'W', 'R', 'I', 'F', 'K', 'J', 'Q', 'M', 'Y', 'O', 'H']
        self.sequence1_2 = ['K', 'Y', 'F', 'M', 'E', 'H', 'B', 'Q', 'D', 'C', 'O', 'J', 'P', 'W', 'V', 'I', 'Z', 'T', 'S', 'G', 'U', 'A', 'L', 'N', 'X', 'R']
        self.sequence2_1 = ['F', 'K', 'H', 'R', 'Y', 'Z', 'U', 'O', 'X', 'V', 'P', 'N', 'M', 'B', 'S', 'C', 'I', 'L', 'J', 'T', 'W', 'E', 'G', 'D', 'A', 'Q']
        self.sequence2_2 = ['X', 'L', 'N', 'D', 'P', 'B', 'Q', 'I', 'T', 'W', 'R', 'Y', 'A', 'V', 'G', 'K', 'E', 'S', 'U', 'C', 'Z', 'M', 'O', 'J', 'H', 'F']
        self.sequence3_1 = ['U', 'N', 'A', 'P', 'V', 'I', 'F', 'D', 'M', 'Y', 'B', 'H', 'R', 'S', 'L', 'J', 'T', 'K', 'Z', 'W', 'G', 'O', 'E', 'Q', 'X', 'C']
        self.sequence3_2 = ['T', 'V', 'D', 'X', 'F', 'M', 'J', 'B', 'K', 'O', 'W', 'R', 'S', 'L', 'N', 'Z', 'G', 'P', 'Q', 'Y', 'E', 'C', 'H', 'U', 'I', 'A']
        self.sequence4_1 = ['M', 'X', 'Q', 'W', 'O', 'D', 'Y', 'H', 'P', 'K', 'U', 'J', 'Z', 'I', 'E', 'F', 'A', 'R', 'L', 'V', 'S', 'T', 'B', 'G', 'C', 'N']
        self.sequence4_2 = ['S', 'L', 'V', 'F', 'C', 'Y', 'D', 'J', 'E', 'U', 'N', 'W', 'G', 'H', 'A', 'P', 'B', 'Q', 'X', 'Z', 'K', 'M', 'T', 'I', 'R', 'O']

        self.roter1_1 = dict(zip(self.sequence1_1, self.sequence1_2))
        self.roter1_2 = dict(zip(self.sequence1_2, self.sequence1_1))
        self.roter2_1 = dict(zip(self.sequence2_1, self.sequence2_2))
        self.roter2_2 = dict(zip(self.sequence2_2, self.sequence2_1))
        self.roter3_1 = dict(zip(self.sequence3_1, self.sequence3_2))
        self.roter3_2 = dict(zip(self.sequence3_2, self.sequence3_1))
        self.reflector1 = dict(zip(self.sequence4_1, self.sequence4_2))
        self.reflector2 = dict(zip(self.sequence4_2, self.sequence4_1))
        # --------------------------------------------------
    
    def setting_values_init(self, setting_code, mode="init"):
        try:
            if mode == "init":
                roter_index1 = self.sequence1_1.index(setting_code[0])
                roter_index2 = self.sequence2_1.index(setting_code[1])
                roter_index3 = self.sequence3_1.index(setting_code[2])
            elif mode == "iter":
                roter_index1 = int(setting_code[0])
                roter_index2 = int(setting_code[1])
                roter_index3 = int(setting_code[2])
            
            for i in range(roter_index1):
                poped = self.sequence1_1.pop(0)
                self.sequence1_1.append(poped)
            
            for i in range(roter_index2):
                poped = self.sequence2_1.pop(0)
                self.sequence2_1.append(poped)
            
            for i in range(roter_index3):
                poped = self.sequence3_1.pop(0)
                self.sequence3_1.append(poped)
            
            self.roter1_1 = dict(zip(self.sequence1_1, self.sequence1_2))
            self.roter1_2 = dict(zip(self.sequence1_2, self.sequence1_1))
            self.roter2_1 = dict(zip(self.sequence2_1, self.sequence2_2))
            self.roter2_2 = dict(zip(self.sequence2_2, self.sequence2_1))
            self.roter3_1 = dict(zip(self.sequence3_1, self.sequence3_2))
            self.roter3_2 = dict(zip(self.sequence3_2, self.sequence3_1))
            self.reflector1 = dict(zip(self.sequence4_1, self.sequence4_2))
            self.reflector2 = dict(zip(self.sequence4_2, self.sequence4_1))
        except:
            print("회전자 설정: 기본값")

    def insert_extra_password(self, outputs):
        if self.order == "encrypt":
            four_code = input("숫자 코드 4자리를 입력하세요(원치 않을 시 Enter): ")
            while True:
                if four_code == None or four_code == "":
                    answer = input("4자리 코드를 입력하지 않을 시 추가적인 보안성 강화 작업이 이루어지지 않습니다. \n그래도 계속 진행하시겠습니까?(y/n): ")
                    if answer == "y" or answer == "Y":
                        print("보안성 강화 작업: 기본값")
                        result = "".join(outputs).lower()
                        return None, result
                    elif answer == "n" or answer == "N":
                        four_code = input("숫자 코드 4자리를 입력하세요(원치 않을 시 Enter): ")
                        continue
                elif len(outputs) < 8:
                    print("보안성 강화에 필요한 접근 코드는 최소 8글자 이상이어야 합니다.")
                    print("보안성 강화 작업: 기본값")
                    result = "".join(outputs).lower()
                    return four_code, result
                
                values = []
                for i, num in enumerate(four_code):
                    outputs.insert(i*2, self.punctuation[int(num)])
                    outputs.insert(-i*4, self.digits[int(num)])
                    values.append(int(num))
                
                outputs.insert(-1, self.punctuation[sum(values)])
                
                result = "".join(outputs).lower()
                return four_code, result
        elif self.order == "decrypt":
            result = "".join(outputs).lower()
            return None, result
    
    def pre_processing_code(self):
        if self.order == "encrypt":
            code = input("접근 코드를 입력하세요(원치 않을 시 Enter): ")
            display = "".join(code)
            
            for word in code:
                if word not in self.lowercase and word not in self.uppercase:
                    code = code.replace(word, "")
            return display, code.upper()
        elif self.order == "decrypt":
            code = input("비밀 번호를 입력하세요(원치 않을 시 Enter): ")
            display = "".join(code)
            extraction = []
            for word in code:
                if word in self.lowercase or word in self.uppercase:
                    extraction.append(word)
            extraction = "".join(extraction)
            return display, extraction.upper()
    
    def display_result(self, result, setting_code, access_code, four_code):
        if self.order == "encrypt":
            print(f"<mode: {self.order}>")
            print("-"*60)
            print(f"your setting code: {setting_code}")
            print("-"*60)
            print(f"your access code: {access_code}")
            print("-"*60)
            print(f"your 4-code: {four_code}")
            print("-"*60)
            print(f"result: {result}")
            print("-"*60)
        elif self.order == "decrypt":
            print(f"<mode: {self.order}>")
            print("-"*60)
            print(f"your setting code: {setting_code}")
            print("-"*60)
            print(f"your password is: {access_code}")
            print("-"*60)
            print(f"result: {result}")
            print("-"*60)

    def generator(self):
        setting_code = input("설정값을 입력하세요(알파벳 대문자 3자리): ")
        self.setting_values_init(setting_code)
        
        display_access_code, access_code = self.pre_processing_code()
        
        input_words_list = []
        output_words_list = []
        
        for word in access_code:
            word = self.plug_codes_dict1[word]
            input_words_list.append(word)
        
        roter_iteration1 = 0
        roter_iteration2 = 0
        roter_iteration3 = 0
        for i in input_words_list:
            forward_word = self.roter3_1[self.roter2_1[self.roter1_1[i]]]
            
            if self.order == "encrypt":
                reflected_word = self.reflector1[forward_word]
            elif self.order == "decrypt":
                reflected_word = self.reflector2[forward_word]
            
            backward_word = self.roter1_2[self.roter2_2[self.roter3_2[reflected_word]]]
            backward_word = self.plug_codes_dict2[backward_word]
            output_words_list.append(backward_word)
            
            roter_iteration1 += 1
            if roter_iteration1 < 26:
                self.setting_values_init("100", "iter")
            elif roter_iteration1 == 26:
                roter_iteration1 = 0
                roter_iteration2 += 1
                self.setting_values_init("110", "iter")
            elif roter_iteration2 == 26:
                roter_iteration2 = 0
                roter_iteration3 += 1
                self.setting_values_init("111", "iter")
        
        four_code, result = self.insert_extra_password(output_words_list)
        
        # 출력부
        self.display_result(result, setting_code, display_access_code, four_code)

password = PasswordEnigma().generator()
# access_code = PasswordEnigma("decrypt").generator()
```

```python
class DocumentEnigma:
    
    def __init__(self, order="encrypt"):
        self.order = order  # 암호생성:encrypt / 암호해독:decrypt
        
        self.punctuation1 = ['<', '}', '%', ';', '~', ']', '^', '"', ',', '/', '>', '\\', '(', '{', '`', '+', '|', '$', '@', '#', '&', '.', '-', '!', '*', ':', ')', '?', '[', '_', '=', "'"]
        self.punctuation2 = [']', '"', '>', '%', '^', '(', ';', '$', "'", '~', '+', '`', ':', '.', '-', '\\', '@', '*', '<', '|', '{', '}', '/', ',', '?', '#', '!', ')', '_', '=', '&', '[']
        self.punctuation_dict1_2 = dict(zip(self.punctuation1, self.punctuation2))
        self.punctuation_dict2_1 = dict(zip(self.punctuation2, self.punctuation1))
        
        self.digits1 = ['3', '6', '5', '2', '8', '0', '4', '9', '1', '7']
        self.digits2 = ['1', '0', '6', '8', '5', '7', '9', '3', '2', '4']
        self.digits_dict1_2 = dict(zip(self.digits1, self.digits2))
        self.digits_dict2_1 = dict(zip(self.digits2, self.digits1))
        
        self.whitespace1 = ['\x0b', '\r', '\t', '\n', ' ', '\x0c']
        self.whitespace2 = [' ', '\n', '\x0c', '\r', '\x0b', '\t']
        self.whitespace1_2 = dict(zip(self.whitespace1, self.whitespace2))
        self.whitespace2_1 = dict(zip(self.whitespace2, self.whitespace1))
        
        self.lowercase = "abcdefghijklmnopqrstuvwxyz"
        self.uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        self.punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
        self.digits = "0123456789"
        self.whitespace = " \t\n\r\x0b\x0c"
        self.printable = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~ \t\n\r\x0b\x0c'
        
        # 플러그 초기화
        # --------------------------------------------------
        left_plugs = ['Y', 'B', 'L', 'U', 'Q', 'S', 'M', 'W', 'Z', 'R', 'D', 'V', 'N', 'G', 'P', 'A', 'H', 'J', 'T', 'I', 'C', 'F', 'K', 'O', 'E', 'X']
        right_plugs = ['W', 'C', 'G', 'H', 'B', 'D', 'V', 'M', 'S', 'I', 'X', 'O', 'E', 'A', 'R', 'U', 'N', 'F', 'P', 'L', 'T', 'Z', 'Y', 'K', 'Q', 'J']
        self.plug_codes_dict1 = dict(zip(left_plugs, right_plugs))
        self.plug_codes_dict2 = dict(zip(right_plugs, left_plugs))
        # --------------------------------------------------
        
        # 회전자 초기화
        # --------------------------------------------------
        self.sequence1_1 = ['B', 'P', 'X', 'Z', 'S', 'G', 'T', 'E', 'L', 'D', 'U', 'A', 'V', 'N', 'C', 'W', 'R', 'I', 'F', 'K', 'J', 'Q', 'M', 'Y', 'O', 'H']
        self.sequence1_2 = ['K', 'Y', 'F', 'M', 'E', 'H', 'B', 'Q', 'D', 'C', 'O', 'J', 'P', 'W', 'V', 'I', 'Z', 'T', 'S', 'G', 'U', 'A', 'L', 'N', 'X', 'R']
        self.sequence2_1 = ['F', 'K', 'H', 'R', 'Y', 'Z', 'U', 'O', 'X', 'V', 'P', 'N', 'M', 'B', 'S', 'C', 'I', 'L', 'J', 'T', 'W', 'E', 'G', 'D', 'A', 'Q']
        self.sequence2_2 = ['X', 'L', 'N', 'D', 'P', 'B', 'Q', 'I', 'T', 'W', 'R', 'Y', 'A', 'V', 'G', 'K', 'E', 'S', 'U', 'C', 'Z', 'M', 'O', 'J', 'H', 'F']
        self.sequence3_1 = ['U', 'N', 'A', 'P', 'V', 'I', 'F', 'D', 'M', 'Y', 'B', 'H', 'R', 'S', 'L', 'J', 'T', 'K', 'Z', 'W', 'G', 'O', 'E', 'Q', 'X', 'C']
        self.sequence3_2 = ['T', 'V', 'D', 'X', 'F', 'M', 'J', 'B', 'K', 'O', 'W', 'R', 'S', 'L', 'N', 'Z', 'G', 'P', 'Q', 'Y', 'E', 'C', 'H', 'U', 'I', 'A']
        self.sequence4_1 = ['M', 'X', 'Q', 'W', 'O', 'D', 'Y', 'H', 'P', 'K', 'U', 'J', 'Z', 'I', 'E', 'F', 'A', 'R', 'L', 'V', 'S', 'T', 'B', 'G', 'C', 'N']
        self.sequence4_2 = ['S', 'L', 'V', 'F', 'C', 'Y', 'D', 'J', 'E', 'U', 'N', 'W', 'G', 'H', 'A', 'P', 'B', 'Q', 'X', 'Z', 'K', 'M', 'T', 'I', 'R', 'O']

        self.roter1_1 = dict(zip(self.sequence1_1, self.sequence1_2))
        self.roter1_2 = dict(zip(self.sequence1_2, self.sequence1_1))
        self.roter2_1 = dict(zip(self.sequence2_1, self.sequence2_2))
        self.roter2_2 = dict(zip(self.sequence2_2, self.sequence2_1))
        self.roter3_1 = dict(zip(self.sequence3_1, self.sequence3_2))
        self.roter3_2 = dict(zip(self.sequence3_2, self.sequence3_1))
        self.reflector1 = dict(zip(self.sequence4_1, self.sequence4_2))
        self.reflector2 = dict(zip(self.sequence4_2, self.sequence4_1))
        # --------------------------------------------------
    
    def setting_values_init(self, setting_code, mode="init"):
        try:
            if mode == "init":
                roter_index1 = self.sequence1_1.index(setting_code[0])
                roter_index2 = self.sequence2_1.index(setting_code[1])
                roter_index3 = self.sequence3_1.index(setting_code[2])
            elif mode == "iter":
                roter_index1 = int(setting_code[0])
                roter_index2 = int(setting_code[1])
                roter_index3 = int(setting_code[2])
            
            for i in range(roter_index1):
                poped = self.sequence1_1.pop(0)
                self.sequence1_1.append(poped)
            
            for i in range(roter_index2):
                poped = self.sequence2_1.pop(0)
                self.sequence2_1.append(poped)
            
            for i in range(roter_index3):
                poped = self.sequence3_1.pop(0)
                self.sequence3_1.append(poped)
            
            self.roter1_1 = dict(zip(self.sequence1_1, self.sequence1_2))
            self.roter1_2 = dict(zip(self.sequence1_2, self.sequence1_1))
            self.roter2_1 = dict(zip(self.sequence2_1, self.sequence2_2))
            self.roter2_2 = dict(zip(self.sequence2_2, self.sequence2_1))
            self.roter3_1 = dict(zip(self.sequence3_1, self.sequence3_2))
            self.roter3_2 = dict(zip(self.sequence3_2, self.sequence3_1))
            self.reflector1 = dict(zip(self.sequence4_1, self.sequence4_2))
            self.reflector2 = dict(zip(self.sequence4_2, self.sequence4_1))
        except:
            print("회전자 설정: 기본값")

    def processing_non_words(self, outputs, non_words):
        if self.order == "encrypt":
            for word in non_words:
                if word[1] in self.punctuation:
                    outputs.insert(word[0], self.punctuation_dict1_2[word[1]])
                elif word[1] in self.digits:
                    outputs.insert(word[0], self.digits_dict1_2[word[1]])
                elif word[1] in self.whitespace:
                    outputs.insert(word[0], self.whitespace1_2[word[1]])
                elif word[1] not in self.printable:
                    outputs.insert(word[0], word[1])
        elif self.order == "decrypt":
            for word in non_words:
                if word[1] in self.punctuation:
                    outputs.insert(word[0], self.punctuation_dict2_1[word[1]])
                elif word[1] in self.digits:
                    outputs.insert(word[0], self.digits_dict2_1[word[1]])
                elif word[1] in self.whitespace:
                    outputs.insert(word[0], self.whitespace2_1[word[1]])
                elif word[1] not in self.printable:
                    outputs.insert(word[0], word[1])
        result = "".join(outputs).lower()
        return result
    
    def pre_processing_code(self):
        if self.order == "encrypt":
            code = input("문서 내용을 입력하세요(원치 않을 시 Enter): ")
        elif self.order == "decrypt":
            code = input("암호화된 문서 내용을 입력하세요(원치 않을 시 Enter): ")
        
        display = "".join(code)
        non_words = []
        for index, word in enumerate(code):
            if word not in self.lowercase and word not in self.uppercase:
                non_words.append([index ,word])
                code = code.replace(word, "")
        return display, non_words, code.upper()
    
    def display_result(self, result, setting_code, contents):
        if self.order == "encrypt":
            print(f"<mode: {self.order}>")
            print("-"*60)
            print(f"your setting code: {setting_code}")
            print("-"*60)
            print(f"your access code: {contents}")
            print("-"*60)
            print(f"result: {result}")
            print("-"*60)
        elif self.order == "decrypt":
            print(f"<mode: {self.order}>")
            print("-"*60)
            print(f"your setting code: {setting_code}")
            print("-"*60)
            print(f"your password is: {contents}")
            print("-"*60)
            print(f"result: {result}")
            print("-"*60)

    def generator(self):
        setting_code = input("설정값을 입력하세요(알파벳 대문자 3자리): ")
        self.setting_values_init(setting_code)
        
        display_contents, non_words, contents = self.pre_processing_code()
        
        input_words_list = []
        output_words_list = []
        
        for word in contents:
            word = self.plug_codes_dict1[word]
            input_words_list.append(word)
        
        roter_iteration1 = 0
        roter_iteration2 = 0
        roter_iteration3 = 0
        for i in input_words_list:
            forward_word = self.roter3_1[self.roter2_1[self.roter1_1[i]]]
            
            if self.order == "encrypt":
                reflected_word = self.reflector1[forward_word]
            elif self.order == "decrypt":
                reflected_word = self.reflector2[forward_word]
            
            backward_word = self.roter1_2[self.roter2_2[self.roter3_2[reflected_word]]]
            backward_word = self.plug_codes_dict2[backward_word]
            output_words_list.append(backward_word)
            
            roter_iteration1 += 1
            if roter_iteration1 < 26:
                self.setting_values_init("100", "iter")
            elif roter_iteration1 == 26:
                roter_iteration1 = 0
                roter_iteration2 += 1
                self.setting_values_init("110", "iter")
            elif roter_iteration2 == 26:
                roter_iteration2 = 0
                roter_iteration3 += 1
                self.setting_values_init("111", "iter")
        
        result = self.processing_non_words(output_words_list, non_words)
        
        # 출력부
        self.display_result(result, setting_code, display_contents)

encrypted_document = DocumentEnigma("decrypt").generator()
```
