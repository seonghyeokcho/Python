```python
import matplotlib.pyplot as plt
import cv2

# 이미지 읽어 들이기
img = cv2.imread("test.jpg")
# 색공간을 그레이스케일로 변환하기
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 역치 지정하기
th = 90
img[img > th] = 255
img[img < th] = 0
# 이미지 출력하기
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# 이미지 읽어 들이기
img = cv2.imread("test.jpg")
# 이미지의 엣지 검출
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.Canny(img, 100, 200, 255)
img = 255 - img
# 이미지 저장하기
cv2.imwrite("out.png", img)

# 이미지 출력하기
#plt.axis("off")
plt.imshow(img, cmap="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
# 이미지 읽어 들이기
img = cv2.imread("test.jpg")
# 원래 이미지를 왼쪽에 출력하기
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# 이미지 좌우 반전하기
plt.subplot(1, 2, 2)
img2 = cv2.flip(img, 0)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
# 이미지 읽어 들이기
img = cv2.imread("test.jpg")
# 이미지 크기 추출하기
h, w, colors = img.shape
size = (w, h)
# 이미지의 중심 위치 추출하기
center = (w // 2, h // 2)

# 회전 변환 행렬 만들기
angle = 45
scale = 1.0
matrix = cv2.getRotationMatrix2D(center, angle, scale)
# 변환하기
img2 = cv2.warpAffine(img, matrix, size)

# 원래 이미지를 왼쪽에 출력하기
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# 회전한 이미지를 오른쪽에 출력하기
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
