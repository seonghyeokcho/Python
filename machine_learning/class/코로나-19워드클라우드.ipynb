image resize

```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("images/heart.png")

img2 = cv2.resize(img, (1000,800))  # 이미지 크기 변경
## 크기 변경은 1번 매개변수는 읽어 들인 이미지, 2번 매개변수는 변경할 크기를 튜플로 지정
## ex) img = cv2.resize(img, (width, height))
cv2.imwrite("images/heart-resize.png", img2)  # 변경한 이미지 저장

plt.axis("off")
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
