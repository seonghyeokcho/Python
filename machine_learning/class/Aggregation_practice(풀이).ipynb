```python
import pandas as pd
```

```python
reserve_tb = pd.read_csv('hotel_data/reserve.csv')
reserve_tb.head()
```

```python
# groupby로 hotel_id를 집약 단위로 설정하여,、
# customer_id의 값을 nunique 함수를 이용하여 고객 수를 카운트
reserve_tb.groupby('hotel_id').agg({'customer_id':'nunique'}).reset_index()
#reserve_tb.groupby('hotel_id')['customer_id'].nunique()
#reserve_tb.groupby('hotel_id')['customer_id'].agg('nunique')
```

```python
# agg 함수를 이용하여 집약 처리를 한꺼번에 지정
# reserve_id를 대상으로 count 함수를 적용
# customer_id를 대상으로 nunique 함수를 적용
reserve_tb.groupby('hotel_id').agg({'reserve_id':'count', 'customer_id':'nunique'})
#reserve_tb.groupby('hotel_id')['reserve_id','customer_id'].agg(['count','nunique'])
```

```python
# 집약 단위를 hotel_id와 people_num의 조합으로 지정
# 집약한 데이터에서 total_price를 추출하여, sum함수에 적용하여 매출 합계 금액을 산출
#reserve_tb.groupby(['hotel_id','people_num'])['total_price'].sum().reset_index()
#reserve_tb.groupby(['hotel_id','people_num']).agg({'total_price':'sum'}).reset_index()
reserve_tb.groupby(['hotel_id','people_num'])['total_price'].agg('sum').reset_index()
```
