```python
import pandas as pd
import glob
```

```python
dt_list = []
for filename in glob.glob('data/*.csv'):
    dt_list.append(pd.read_csv(filename))
```

## college.csv : 미국 대학 데이터

```python
dt_list[1].head()
```

- INSTNM : 대학교명
- CITY : 대학이 위치한 도시명
- STABBR : 대학이 위치한 주(state)명
- HBCU : 1964년 인권법 제정 이전 흑인들이 모여 있는 지역에 세워진 고등교육기관 연합 여부 (1: True, 2:False)
- MENONLY : 남자대학 여부 (1: True, 0:False)
- WOMENONLY : 여자대학 여부 (1: True, 0:False)
- RELAFFIL : 종교대학여부 (1: True, 0:False)
- SATVRMID : SAT 구술 점수 중앙값
- SATMTMID : SAT 수학 점수 중앙값
- DISTANCEONLY : 원격교육만 제공하는 대학(사이버대학 등) 여부(1:True, 0:False)
- UGDS : 학부생 수
- UGDS_WHITE : 학부생 중 백인학생 비율
- UGDS_BLACK : 학부생 중 흑인학생 비율
- UGDS_HISP : 학부생 중 히스패닉 학생 비율
- UGDS_ASIAN : 학부생 중 아시아 학생 비율
- UGDS_AIAN : 학부생 중 미국인디언/알래스카원주민 학생 비율
- UGDS_NHPI : 학부생 중 하와이/환태평양 지역 학생 비율
- UGDS_2MOR : 학부생 중 2개인종이상 혼혈 학생 비율
- UGDS_NRA : 학부생 중 비이민 신분 학생(유학생) 비율
- UGDS_UNKN :인종을 알수 없는 학생 비율
- PPTUG_EF : 파트타임 학생 비율
- CURROPER : 현재 운영 중인지 여부(1:True, 0: False)
- PCTPELL : 학부생 중 Pell Grant(연방 장학금) 학생 비율
- PCTFLOAN : 학부생 중 연방학자금 대출 학생 비율
- UG25ABV : 25살 이상의 학부생 비율
- MD_EARN_WNE_P10 : 입학 후 10년간 수입 중앙값 - 공개하지 않은 경우 PrivacySuppressed 로 표시
- GRAD_DEBT_MDN_SUPP : 학자금 대출 완납에 소요된 시간(일) 중앙값 - 공개하지 않은 경우 PrivacySuppressed 로 표시

- MONTH : 비행 월
- DAY : 비행 일

     - EV:	Atlantic Southeast Airlines

     - MQ:	American Eagle Airlines Inc.
     - VX:	Virgin America

- DEST_AIR : 도착공항
