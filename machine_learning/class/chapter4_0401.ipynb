## 사전 정렬하기(sort)

```python
names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 
'Michale':27115, 'Bob':5887, 'Kelly':7855}
ret1 = sorted(names)
print(ret1)
ret = sorted(names.values())
print(ret)

def f1(x):
   return x[0]

def f2(x):
   return x[1]

ret2 = sorted(names.items(), key=f1) # 키를 기준으로 정렬
print(ret2)

ret3 = sorted(names.items(), key=f2) # 값을 기준으로 정렬
print(ret3)

ret4 = sorted(names.items(), key=f2, reverse=True)
print(ret4)
```

## 인자를 바꾸면서 함수 반복 호출

```python
# 파이썬 내장함수 map()은 집합 A와 함수 f가 주어지면
# 집함 B를 구해주는 함수와 동일한 역할 수행

f = lambda x: x**2
#print(f(3))
args = [1,2,3,4,5]
print(list(map(f,args)))
```

```python
f = lambda a,b,c,d,e,f,g,h,i,j,k: (a*a)+(c-d)/(e**e)+(f-g)*(i**i)+(j-k)
a = [1,2,3,4,5]
b = [1,2,3,4,5]
c = [1,2,3,4,5]
d = [1,2,3,4,5]
e = [1,2,3,4,5]
ff = [1,2,3,4,5]
g = [1,2,3,4,5]
h = [1,2,3,4,5]
i = [1,2,3,4,5]
j = [1,2,3,4,5]
k = [1,2,3,4,5]
print(list(map(f,a,b,c,d,e,ff,g,h,i,j,k)))
```

## 날짜 

```python
# 현재 시간 출력하기
from time import localtime, strftime

logfile = 'test.log'
def writelog(logfile, log):
    time_stamp = strftime('%Y-%m-%d %X\t', localtime())
    log = time_stamp + log + '\t'
    
    with open(logfile, 'a') as f:
        f.writelines(log)
```

```python
writelog(logfile, '첫 번째 로깅 문장입니다.')
```

```python
# 올해 경과된 날짜수 계산하기
t = localtime()
start_day = '%d-01-01' %t.tm_year
elapsed_day = t.tm_yday
print('오늘은 [%s]이후 [%d]일째 되는 날입니다.' %(start_day, elapsed_day))
```

```python
# 오늘의 요일 계산하기
weekdays = ['월요일','화요일','수요일','목요일','금요일','토요일','일요일']

t = localtime()
today = '%d-%d-%d' %(t.tm_year, t.tm_mon, t.tm_mday)
week = weekdays[t.tm_wday]

print('[%s] 오늘은 [%s] 입니다.'%(today,week))
```
