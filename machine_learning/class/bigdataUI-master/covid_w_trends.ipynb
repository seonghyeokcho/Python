```python
from bs4 import BeautifulSoup
import requests
from pandas import Series, DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

res = requests.get("http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=14&ncvContSeq=&contSeq=&board_id=&gubun=")
res = res.text
res = BeautifulSoup(res, 'html.parser')

k = 0

date = []
lst = res.find_all('th',{'scope':'col'})

for i in lst:
    date.append(i.text)

date_1 = date[3:10]        

country = []
lst1 = res.find_all('th', {'scope':'row'})

for j in lst1:
    country.append(j.text)

country_1 = country[8:27]   


table = res.find('table', {'class': 'num minisize'})

data = []

for a in table.find_all("tr"):
    infolist = []

    for b in a.find_all("td"):
        info = b.get_text()
        infolist.append(info)

    if len(infolist) > 0 :
        data.append(infolist)

date = date[3:11]

datas = list(zip(date,*data))
df = pd.DataFrame(datas,columns=["날짜"]+country_1).T
df.to_csv("data/wdtd_covid.csv")  # 저장하고 싶은 위치로 경로 설정하면 크롤링된 데이터가 csv형식으로 만들어짐
```

```python
data = pd.read_csv('data/wdtd_covid1.csv',index_col=[0])
data
```

```python
melteddata = data.melt(id_vars=['날짜'], value_vars=['미국','스페인','이탈리아','독일','영국','프랑스','터키','이란',
                                      '중국','벨기에','캐나다','네덜란드','스위스','일본','한국',
                                      '인도네시아','필리핀','태국','베트남'], var_name='국가', value_name='현황')
melteddata.head(10)
```

```python
def split_data(data):
    
    try:
        data.loc[:,'현황'].replace(''," ")
    except Exception as e:
        raise e
    return data;
```

```python
split_data(melteddata)
```

```python
pivot_melted = melteddata.pivot(index='날짜',columns='국가', values='현황')
pivot_melted
```

```python
pivot_melted.to_csv('melted_data_covid.csv',encoding='cp949')
```
