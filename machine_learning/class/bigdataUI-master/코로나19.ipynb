```python
from bs4 import BeautifulSoup
import requests
from pandas import Series, DataFrame
import numpy as np
import pandas as pd

res = requests.get("http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=13&ncvContSeq=&contSeq=&board_id=&gubun=")
res = res.text
res = BeautifulSoup(res, 'html.parser')



city = []
cases = []
add = []
lst = res.find_all('span',{'class':'name'})
for i in lst:
    city.append(i.text)

lst2 = res.find_all('span',{'class':'num'})
for j in lst2:
    cases.append(j.text)

lst3 = res.find_all('span',{'class':'before'})

for k in lst3:
    add.append(k.text)

```

```python
city
```

```python
df = DataFrame(city) 
df.columns=["Korea_city"]
df
```

```python
df1 = DataFrame(cases)
df1.columns=["Sad_Cases"]
df1.head(18)
```

```python
df2 = DataFrame(add)
df2.columns=["Add_cases"]
df2.head(18)
```

```python
result1 = pd.concat([df,df1,df2],axis=1,join='inner')
result1
```
